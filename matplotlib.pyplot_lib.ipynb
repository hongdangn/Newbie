# Pyplot API and Object-Oriented (OO) API

## Pyplot API

```python
x = np.linspace(1, 10, 1000)
plt.plot(x, np.cos(x), color = "blue", label = "cos(x)")
plt.plot(x, np.sin(x), color = "red", label = "sin(x)")

plt.title("sin(x) and cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")

#plt.xlim(1, 5)
#plt.ylim(0.1, 0.7)
#plt.axis([1, 5, 0.1, 0.7])
plt.axis("tight")
plt.axis("equal")
plt.legend();
```

## Object-Oriented API

```python
#1. Prepare data
x = np.linspace(1, 4, 4).astype(int)
y = np.linspace(11, 44, 4).astype(int)

#2. Set up plot
fig, ax = plt.subplots() # figure size = Width and Height of the plot

#3. Plot the data
ax.plot(x, y, color = "blue");
ax.set(title = "the plot", xlabel = "x", ylabel = "y");
```

## Line

```python
# Array
x = np.linspace(0, 10, 100)
x[:5]
```

```python
# API plot line chart
fig, ax = plt.subplots()
ax.plot(x, x**3);
```

## Scatter

```python
# Pyplot API
plt.scatter(x, x**3);
```

```python
# OO API

x = np.random.random(size = 100)
y = np.random.random(size = 100)

sizes = 200*x
colors = np.random.rand(100)

fig, ax = plt.subplots()
img = ax.scatter(x, y, s = sizes, c = colors, cmap = 'rainbow');

fig.colorbar(img);
```

## Bar

* Vertical
* Horizontal

```python
# Vertical

# Prepare data

drink_prices = {"Coke": 10, "Pepsi": 12, "Spring": 14}
fig, ax = plt.subplots()

ax.bar(drink_prices.keys(), drink_prices.values());
ax.set(title = "Drink prices", xlabel = "Drink", ylabel = "Price ($)");
```

```python
# Horizontal

fig, ax = plt.subplots()
ax.barh(list(drink_prices.keys()), list(drink_prices.values()));
```

## Histogram

```python
# Prepare data

student_height = np.random.normal(170, 10, 250)
student_height
```

```python
student_height[:10]
```

```python
# pyplot API

plt.hist(student_height, bins = 30);
```

# Subplots

```python
# Option 1:

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = [10, 5])

# plot data for each axis
ax1.plot(x, x + 1);
ax1.plot(x, x/2);

ax2.scatter(np.random.random(10), np.random.randint(1, 10, 10));
ax3.bar(drink_prices.keys(), drink_prices.values());
ax4.hist(student_height, bins = 30);
```

```python
# Option 2:

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = [10, 5])

# plot data for each axis
ax[0, 0].plot(x, x + 1);
ax[0, 0].plot(x, x/2);

ax[0, 1].scatter(np.random.random(10), np.random.randint(1, 10, 10));
ax[1, 0].bar(drink_prices.keys(), drink_prices.values());
ax[1, 1].hist(student_height, bins = 30);
```
