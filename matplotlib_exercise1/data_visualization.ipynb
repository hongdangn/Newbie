```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available
plt.style.use("seaborn-v0_8-whitegrid")
```

```python
cities = pd.read_csv("california_cities.csv")
cities.head()

cities["area_total_km2"].sort_values()
```

```python
# extract latitude(latd) and longtitude(longd)

latd, longd = cities["latd"], cities["longd"]
population, area = cities["population_total"], cities["area_total_km2"]

# plot using pyplot API

# create legend for cities's sizes

sizes = [50, 100, 200, 500]

for item in sizes:
    plt.scatter([], [], s = item, c = 'blue', label = str(item) + "km$^2$")
    
plt.legend(title = "Area", labelspacing = 1)

plt.scatter(longd, latd, c = np.log10(population), s = area, cmap = 'rainbow');
plt.axis("equal");

plt.xlabel("Longtitude");
plt.ylabel("Latitude");

plt.colorbar()
plt.clim(3, 7)
plt.title("California city: Population and Area distribution");
```

```python
# plot using OO API

fig, ax = plt.subplots()
plot = ax.scatter(longd, latd, c = np.log10(population), s = area, cmap = "cool");

for item in sizes:
    ax.scatter([], [], s = item, c = "blue", label = str(item) + "km$^2$")
    
ax.legend(labelspacing = 1)

fig.colorbar(plot);
```
