```python
# Remove the two extra zeros at the end and the '$' of the price column

car_sales["Price"] = car_sales["Price"].apply(lambda x: x.replace("$", "").replace(",", "").replace(".", ""))
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].apply(lambda x: int(x))
car_sales.info()
```

```python
# Create a column which represents the price of a car per kilometer
# Then view the DataFrame

car_sales["Price_per_KM"] = car_sales["Price"]/car_sales["Odometer (KM)"] 
car_sales
```
